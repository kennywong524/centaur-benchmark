```python
import subprocess, sys

def run(*args: str) -> None:
    cmd = [sys.executable, "-m", "centaur_benchmark", *args]
    print(" ".join(cmd))
    subprocess.run(cmd, check=True)

# Example (uncomment after `centaur-benchmark login`):
# run("run", "--task", "meal_plan", "--mode", "augmentation", "--assistants", "gpt-4.1")
# run("judge", "--task", "meal_plan", "--run", "<RUN_ID>", "--subset", "augmentation")
# run("summarize", "--task", "meal_plan", "--run", "<RUN_ID>")
```
